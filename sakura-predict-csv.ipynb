## データの加工
予測に有用な特徴量を作成する。

```python
# セル3-1. ループ処理用のリスト作成

# location_nameとyearの一覧を取得する
location_list = df_weather['location_name'].drop_duplicates().values
year_list = df_weather['year'].drop_duplicates().values

print(location_list)
print(year_list)
```

```python
# セル3-2. 特徴量の作成

from datetime import datetime

print('Feature extraction is started at ', datetime.now().strftime("%Y/%m/%d %H:%M:%S"))

# 加工後のデータを格納するための空DataFrameを作成しておく
df_weather_custom = pd.DataFrame(index=[], columns=[])

for current_location in location_list:
    for current_year in year_list:
        df_one = df_weather[(df_weather.location_name == current_location) & (df_weather.year == current_year)].copy()

        # NULL値（NaN）を0へ置換する。降水量（precipitation）に関して、0mmの降雨と降雨なし（Null）は異なるが、今回は同じとみなす
        df_one = df_one.fillna(0)

        # 「3日後に開花しているかどうか」をラベル（モデルが出力すべき正解の値）として定義する
        df_one['label'] = df_one['blooming'].shift(-3)

        # データ期間の終わりである6/27 - 6/30は「3日後のデータ」が存在せず、NaNとなるので1（開花）に変更する
        df_one['label'] = df_one['label'].fillna(1)

        # 気温0℃以下を一律0℃とみなす
        df_one['temperature_highest'] = df_one['temperature_highest'].where(df_one['temperature_highest'] >=0,0)
        df_one['temperature_avg'] = df_one['temperature_avg'].where(df_one['temperature_avg'] >=0,0)

        # 最高気温の7日間移動平均を算出する
        df_one['th_mean7'] = df_one['temperature_highest'].rolling(window=7, min_periods=1).mean()

        # 2月以降、3月以降の平均気温の累積値をそれぞれ算出する
        df_one['ta_cumsum2'] = df_one['temperature_avg'].where(df_one.index.month >= 2,0).cumsum()
        df_one['ta_cumsum3'] = df_one['temperature_avg'].where(df_one.index.month >= 3,0).cumsum()
        
        # 気温3℃以下を一律3℃とみなす
        df_one['temperature_highest'] = df_one['temperature_highest'].where(df_one['temperature_highest'] >=3,3)
        df_one['temperature_avg'] = df_one['temperature_avg'].where(df_one['temperature_avg'] >=3,3)
        
        # 1月の平均気温の平均を算出する
        df_one['ta_mean1'] = df_one['temperature_avg'].where(df_one.index.month == 1,0).mean()
        

        df_weather_custom = pd.concat([df_weather_custom, df_one])

print('Feature extraction is completed at ', datetime.now().strftime("%Y/%m/%d %H:%M:%S"))
```

## モデル作成
機械学習のライブラリーを利用して予測モデルを作成する。

```python
# セル4-1. ライブラリーのインポート

# 今回はscikit-learnを使用する
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
# セル4-2. モデルの定義と学習

#ラベルと説明変数をNumpy Arrayとして用意
label = np.array(df_weather_custom['label'])
features = np.array(df_weather_custom[["ta_mean1", "th_mean7", "ta_cumsum2", "ta_cumsum3"]])


# 2016年以前のデータを学習用に、2017年のデータをテスト用に分割
train_index = np.array(df_weather_custom['year'] <= 2016)
test_index = np.array(df_weather_custom['year'] > 2016)

# スケールの異なる説明変数を含むため標準化を行う
sc = StandardScaler()
sc.fit(features[train_index])
features = sc.transform(features)

#ロジスティック回帰モデルを学習
classifier = LogisticRegression(C=1.0, penalty='l2')
classifier.fit(features[train_index], label[train_index]) 
```

```python
# セル4-3. テストデータでの予測
df_prediction = df_weather_custom[test_index]
# SettingWithCopyWarningを出さないための設定
df_prediction.is_copy = False
df_prediction['prediction'] = classifier.predict(features[test_index])
```

```python
# セル4-4. 予測結果の表示
df_prediction[df_prediction.location_name == '奈良'][['date', 'location_name', 'label', 'prediction']].iloc[86:91]
```

```python
# セル4-5. モデルの評価

# 実測値（label）および予測結果（prediction）が開花（1）になっている日付をそれぞれ取得する
label2017 = df_prediction[df_prediction.label == 1].groupby(['location_name']).min()['date'].astype('datetime64[ns]')
prediction2017 = df_prediction[df_prediction.prediction == 1].groupby(['location_name']).min()['date'].astype('datetime64[ns]')

# 観測所名（location_name）をIndexとして実測値と予測値の差を計算する
prediction_summary = pd.concat([label2017,prediction2017],axis=1)
prediction_summary.columns = ['label2017', 'prediction2017']
prediction_summary['diff2017'] = pd.to_timedelta(prediction_summary['label2017'] -  prediction_summary['prediction2017'])
prediction_summary['diff2017_days'] = (prediction_summary['diff2017'] / np.timedelta64(1, 'D')).astype(int)
prediction_summary
```

```python
# セル4-6. 地図上での誤差の表示

# 予測の誤差と観測所の位置情報を結合する
prediction_summary = pd.concat([prediction_summary, df_station], axis=1)
prediction_summary.head()

# 地図上に予測の誤差を表示する
%brunel map('japan') + data('prediction_summary') x(longitude) y(latitude) \
tooltip(location_name,label2017,prediction2017) color(diff2017_days) \
label(diff2017_days) ::width=1200, height=600 
```
